### 1.6.2 pyplot functions

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-4, 2, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(x)

plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--', label='cos')
plt.plot(x, y3, color='m', label='exponential')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos & exponential')
plt.legend()
plt.show()
```
